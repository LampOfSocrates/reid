This notebook uses `coursework.src.utils.explore.show_veri_good_and_junk` to inspect a query image together with its good and junk matches.

- Colab data root: `/content/drive/MyDrive/Veri`
- Local data root: `G:\\My Drive\\VeRi`

```python
from pathlib import Path
import sys


def is_colab() -> bool:
    try:
        import google.colab  # type: ignore
        return True
    except ImportError:
        return False


if is_colab():
    repo_root = Path('/content/reid')
else:
    repo_root = Path.cwd()

if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

print(f'is_colab={is_colab()}')
print(f'repo_root={repo_root}')
```

```python
from pathlib import Path

DATA_ROOT = Path('/content/drive/MyDrive/Veri') if is_colab() else Path(r'G:\\My Drive\\VeRi')
QUERY_INDEX = 0
INDEX_IS_ONE_BASED = False
GALLERY_INDICES_ARE_ONE_BASED = True
MAX_GOOD = 10
MAX_JUNK = 10
FIGSIZE = (36, 16)

print(f'Data root: {DATA_ROOT}')
print(f'Query index: {QUERY_INDEX}')
```

```python
from coursework.src.utils.explore import show_veri_good_and_junk

assert DATA_ROOT.exists(), f'Missing data root: {DATA_ROOT}'

result = show_veri_good_and_junk(
    root=DATA_ROOT,
    query_index=QUERY_INDEX,
    index_is_one_based=INDEX_IS_ONE_BASED,
    gallery_indices_are_one_based=GALLERY_INDICES_ARE_ONE_BASED,
    max_good=MAX_GOOD,
    max_junk=MAX_JUNK,
    figsize=FIGSIZE,
    show_plot=True,
)

result
```
